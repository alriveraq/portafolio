# Data preparation

##Preparando data frame para trabajar

###Data

```python
data.unstack()
solodata = data.loc['data']
solodata
```

###Metadata

```python
data.unstack()
metadata = data.loc['metadata']
metadata = metadata.reset_index()
metadata = metadata.T
metadata
```

Tras armar nuestro dataframe se detecto que la información esta almacenada en un multi-index, el primer index se llama data y el segundo metadata.

Ambos index tienen información que se complementa una con la otra pero uno cuenta la información y el otro da la información por sectores.

El metadata almacena la informacion total de las votaciones por sesion, esto se detalla mostrando las votaciones totales de la sesion, el tipo de votación, el numero de la sesion, el tema y el resultado.

El data almacena la información de los senadores, contiene cada votacion de los senadores en las diferentes sesiones y  las votaciones pueden ser Si, no, pareo o abstención.

Para trabajar en nuestro proyecto usaremos solo el metadata ya que queremos predecir cuantos Si pueden haber en una sesion dependiendo de los votos, el tipo de votacion y la etapa.

##Creando nuevo dataframe

```python
dft = pd.DataFrame(columns= ['Ab','No','TipoVotacion','Etapa','Si'])
dft['Ab']=metadata[0]
dft['No']=metadata[3]
dft['Si']=metadata[7]
dft['TipoVotacion']=metadata[9]
dft['Etapa']=metadata[1]
dft=dft.drop(dft.index[[0]])

dft
```

Aquí estamos sacando las votaciones totales de las diferentes sesiones registradas hasta la fecha exceptuando los pareo ya que cuentan como 2 votos y podría confundirse al momento de aplicarlo

##Transformacion de varibles

```python
from sklearn.preprocessing import LabelEncoder
Senado = LabelEncoder()
dft['Etapa'] = Senado.fit_transform(dft['Etapa'])
dft['TipoVotacion'] = Senado.fit_transform(dft['TipoVotacion'])
dft
```

En este caso solo transformamos la etapa y el tipo de votacion con el labelencoder, esto hace que los tipo String cambien a un int, con esto podemos someter al dataframe a un entrenamiento para implemantarlos en los modelos.

##Training

```python
import numpy as np
from sklearn.model_selection import train_test_split
X = np.array(dft.drop(['Si'],1))
y = np.array(dft['Si'])
validation_size = 0.20
seed = 7
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

#Modeling

##Regresion Arbol de decisiones


```python
from sklearn import tree

clf = tree.DecisionTreeRegressor()
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
clf.score(X_train,y_train)
```

```python
from sklearn.metrics import max_error
max_error(y_test, y_pred)
```

Con este modelo puede agregar 18 votos "Si" extras en su error maximo

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

En este modelo tiene un 35% de variacion al resultado original
